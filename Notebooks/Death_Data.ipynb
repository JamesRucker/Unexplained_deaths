```python
# Import dependencies
import pandas as pd
```

```python
# Grab relevant death data dataframes and append them, then fill all missing values with 0
death_2014_2018_df = pd.read_csv("../Data/Weekly_Counts_of_Deaths_by_State_and_Select_Causes__2014-2018_studydata.csv")
death_2019_2020_df = pd.read_csv("../Data/Weekly_Counts_of_Deaths_by_State_and_Select_Causes__2019-2020_studydata.csv")
death_2014_2020_df = death_2014_2018_df.append(death_2019_2020_df)
death_2014_2020_df.fillna(0, inplace = True)

# Normalization dataframe is imported
norm_df = pd.read_csv("../Data/normalization_year.csv")
norm_year = norm_df.columns

death_2014_2020_df
```

```python
# Jurisdictions with bad data that need to be isolated and dropped plus the dataframe containing aggregated data are extracted 

# Array of jurisdictions to drop
juri_to_drop = ["North Carolina", "District of Columbia", "Puerto Rico", "Connecticut"]


# For loop adds dataframes of interest to an array where each entry is the dataframe to be examined
state_drop_df = death_2014_2020_df.loc[death_2014_2020_df["jurisdiction"].isin(juri_to_drop) == True]
united_df = death_2014_2020_df.loc[death_2014_2020_df["jurisdiction"]=="United States"]


united_df
```

```python
# Dates to drop, being standardized, are formed into a list

# Dates before the date to drop are made into an initial list
drop = united_df.loc[united_df["week_end_date"]=="10/11/2014"].index[0]-1
dates_to_drop_init = united_df.loc[:drop]["week_end_date"]

# Dates following the desired end date are appended to the list
drop = united_df.loc[united_df["week_end_date"]=="10/3/2020"].index[0]
dates_to_drop = dates_to_drop_init.append(united_df.loc[drop:]["week_end_date"])


dates_to_drop
```

```python
# Reset all indices and drop unwanted dates for relevant dataframes



# Both dataframes are trimmed of unwanted dates.  United also has index reset.
cleaned_united_df = united_df[united_df["week_end_date"].isin(dates_to_drop) == False].reset_index().drop("index", axis=1)
cleaned_state_drop_df = state_drop_df[state_drop_df["week_end_date"].isin(dates_to_drop) == False]

cleaned_united_df
```

```python
# Dataframes of values to be subtracted from United States values is crafted. Note that the indexing changes.
bad_values = cleaned_state_drop_df.groupby("week_end_date").sum().reset_index()
bad_values
```

```python
# Bad Values are subtracted from united states values to give Values without the excluded states
# Note that as the indexing differs, and the indexing must be preserved for the united states dataframes, a for loop going
#     through the rows will be used and a dataframe will be constructed from that.

# Empty dataframe holding final data is initialized. This is done as there is no empty value for jurisdiction.
final_df = cleaned_united_df.loc[cleaned_united_df["jurisdiction"] == ""]

# Count variable for the week in the flu season and the flu season itself are initialized
week = 1
season = 0

# For loop that goes through each row
for index, row in cleaned_united_df.iterrows():

    # Grabs the datum point from the bad_value_df corresponding to the correct date
    bad_datum = bad_values.loc[bad_values["week_end_date"] == row[1]]
    
    # Temporary dataframe holding only the one datum point for that date.
    # Relevant data has bad data subtracted from wanted, then multiplied by the normalization factor
    # Columns for season and week are added for ease later
    temp_df = pd.DataFrame([[row[0],
                             row[1],
                             week,
                             season,
                             float(row[2] - bad_datum["all_cause"]) * float(norm_df[norm_year[season]]),
                             float(row[3] - bad_datum["septicemia"]) * float(norm_df[norm_year[season]]),
                             float(row[4] - bad_datum["cancer"]) * float(norm_df[norm_year[season]]),
                             float(row[5] - bad_datum["diabetes"]) * float(norm_df[norm_year[season]]),
                             float(row[6] - bad_datum["Alzheimers"]) * float(norm_df[norm_year[season]]),
                             float(row[7] - bad_datum["influenza_pneumonia"]) * float(norm_df[norm_year[season]]),
                             float(row[8] - bad_datum["chronic_LRD"]) * float(norm_df[norm_year[season]]),
                             float(row[9] - bad_datum["other_respiratory"]) * float(norm_df[norm_year[season]]),
                             float(row[10] - bad_datum["kidney"]) * float(norm_df[norm_year[season]]),
                             float(row[11] - bad_datum["NEC"]) * float(norm_df[norm_year[season]]),
                             float(row[12] - bad_datum["heart_disease"]) * float(norm_df[norm_year[season]]),
                             float(row[13] - bad_datum["CBVD"]) * float(norm_df[norm_year[season]]),
                             float(row[14] - bad_datum["COVID19_UCOD"]) * float(norm_df[norm_year[season]])
                            ]],
                           columns=["jurisdiction", "week_end_date", "season_week", "season", "all_cause", "septicemia",\
                                    "cancer", "diabetes", "Alzheimers", "influenza_pneumonia", "chronic_LRD",\
                                    "other_respiratory", "kidney", "NEC", "heart_disease", "CBVD", "COVID19_UCOD"
                 ])
    
    # Row for specific data is appended to dataframe
    final_df = final_df.append(temp_df, ignore_index = True)
    
    # Checks are made for week in flu season.  If flu season over, start next one, otherwise add 1 week.
    if week == 52:
        week = 1
        season+=1
    else:
        week+=1

final_df
```

```python
# Export final dataframe to csv
final_df.to_csv("../Data/final_death_data.csv", index=False)
```
