```python
# Import Dependencies
import pandas as pd

# Grab dataframes for averages
aggregated_2019_df = pd.read_csv("../Data/aggregated_2019_final.csv")
aggregated_2_year_df = pd.read_csv("../Data/aggregated_2_year_final.csv")
aggregated_3_year_df = pd.read_csv("../Data/aggregated_3_year_final.csv")
aggregated_5_year_df = pd.read_csv("../Data/aggregated_5_year_final.csv")

# Slices for periods are defined
periods = [slice(16,26,1),slice(26,35,1),slice(35,52,1)]
period_names = ["Pandemic_Awareness", "Initial_Pandemic_Peak", "Pandemic_Post-Peak"]

# Dataframes formed into array
frames = [aggregated_2_year_df, aggregated_3_year_df, aggregated_5_year_df, aggregated_2019_df]

aggregated_2019_df
```

```python
# Dataframes for each period are crafted

# Row names carfted for each of reading
rows = ["Prior-season mean, 2-season benchmark","Prior-season mean, 3-season benchmark",
        "Prior-season mean, 5-season benchmark", "2019-2020", "Change, 2-season benchmark", "Change, 3-season benchmark",
        "Change, 5-season benchmark", "% of change in all-cause deaths, 2-season",
        "% of change in all-cause deaths, 3-season", "% of change in all-cause deaths, 5-season"]

for j in range(3):
    
    # Initial sums
    for i in range(4):
        temp_df = pd.DataFrame([[rows[i],\
                                frames[i].iloc[periods[j]].sum()["all_cause"],
                                frames[i].iloc[periods[j]].sum()["COVID19_UCOD"],
                                frames[i].iloc[periods[j]].sum()["Explained"]-frames[i].iloc[periods[j]].sum()["COVID19_UCOD"],
                                frames[i].iloc[periods[j]].sum()["NEC"],
                                frames[i].iloc[periods[j]].sum()["Total Unreported"],
                                frames[i].iloc[periods[j]].sum()["Unexplained"]]],
                                   columns = ["", "All Causes", "COVID-19", "Report Causes, Not COVID-19", "NEC", "Unreported",
                                             "Total Unexplained"]
                              )

        if i==0:
            holding_df = temp_df
        else:
            holding_df = holding_df.append(temp_df, ignore_index=True)

    # Change for each season
    for i in range(3):
        temp_df = pd.DataFrame([[rows[i+4],\
                                holding_df.iloc[3]["All Causes"]-holding_df.iloc[i]["All Causes"],
                                holding_df.iloc[3]["COVID-19"]-holding_df.iloc[i]["COVID-19"],
                                holding_df.iloc[3]["Report Causes, Not COVID-19"]-holding_df.iloc[i]["Report Causes, Not COVID-19"],
                                holding_df.iloc[3]["NEC"]-holding_df.iloc[i]["NEC"],
                                holding_df.iloc[3]["Unreported"]-holding_df.iloc[i]["Unreported"],
                                holding_df.iloc[3]["Total Unexplained"]-holding_df.iloc[i]["Total Unexplained"]]],
                                   columns = ["", "All Causes", "COVID-19", "Report Causes, Not COVID-19", "NEC",
                                              "Unreported", "Total Unexplained"]
                          )

        holding_df = holding_df.append(temp_df, ignore_index=True)

    # Percent of change of all-cause for each season
    for i in range(3):
        temp_df = pd.DataFrame([[rows[i+7],\
                                holding_df.iloc[i+4]["All Causes"]/holding_df.iloc[i+4]["All Causes"],
                                holding_df.iloc[i+4]["COVID-19"]/holding_df.iloc[i+4]["All Causes"],
                                holding_df.iloc[i+4]["Report Causes, Not COVID-19"]/holding_df.iloc[i+4]["All Causes"],
                                holding_df.iloc[i+4]["NEC"]/holding_df.iloc[i+4]["All Causes"],
                                holding_df.iloc[i+4]["Unreported"]/holding_df.iloc[i+4]["All Causes"],
                                holding_df.iloc[i+4]["Total Unexplained"]/holding_df.iloc[i+4]["All Causes"]]],
                                   columns = ["", "All Causes", "COVID-19", "Report Causes, Not COVID-19", "NEC",
                                              "Unreported", "Total Unexplained"]
                          )

        holding_df = holding_df.append(temp_df, ignore_index=True)

    holding_df.to_csv(f'../Data/Periods_{period_names[j]}_changes.csv', index=False)
    
print(f'Period: {period_names[j]}')
holding_df
```
