```python
# Import dependencies
import pandas as pd

# Grab finished Dataframe
death_df = pd.read_csv("../Data/final_death_data.csv")
death_df
```

```python
# Dataframes for 2019 season and 2-, 3-, and 5-year average before 2019 season are crafted then exported

# Arrays to grab years of interest.  Season week 0 corresponds to 2014, with incrementing values incrementing year.
season_numb = [[5], [3,4], [2,3,4], [0,1,2,3,4]]
title = ["2019_season", "2_year_average", "3_year_average", "5_year_average"]

# Loop through all relevant averages
for i in range(4):
    
    # Data for relevant seasons are extracted
    data_df = death_df.loc[death_df["season"].isin(season_numb[i])].drop("season", axis=1)

    # Data is grouped by season_week, averaged, and the resulting dataframe is exported to csv
    average_df = data_df.groupby("season_week").mean()
    average_df.to_csv(f'../Data/Aggregated_{title[i]}.csv')
```
