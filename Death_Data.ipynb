```python
# Import dependencies
import pandas as pd
```

```python
# Grab relevant dataframes and append them
death_2014_2018_df = pd.read_csv("data/Weekly_Counts_of_Deaths_by_State_and_Select_Causes__2014-2018_studydata.csv")
death_2019_2020_df = pd.read_csv("data/Weekly_Counts_of_Deaths_by_State_and_Select_Causes__2019-2020_studydata.csv")
death_2014_2020_df = death_2014_2018_df.append(death_2019_2020_df)


death_2014_2020_df
```

```python
# Jurisdictions with bad data that need to be isolated and dropped plus the dataframe containing aggregated data are extracted 

# Array of jurisdictions to drop
juri_interest = ["United States", "North Carolina", "District of Columbia", "Puerto Rico", "Connecticut"]

# Empty array to hold dataframes is defined
state_interest_df = []

# For loop adds dataframes of interest to an array where each entry is the dataframe to be examined
for state in juri_interest:
    state_interest_df.append(death_2014_2020_df.loc[death_2014_2020_df["jurisdiction"]==state])


state_interest_df[0]
```

```python
# Dates to drop, being standardized, are formed into a list

# Dates before the date to drop are made into an initial list
drop = state_interest_df[0].loc[state_interest_df[0]["week_end_date"]=="10/11/2014"].index[0]-1
dates_to_drop_init = state_interest_df[0].loc[:drop]["week_end_date"]

# Dates following the desired end date are appended to the list
drop = state_interest_df[0].loc[state_interest_df[0]["week_end_date"]=="10/3/2020"].index[0]
dates_to_drop = dates_to_drop_init.append(state_interest_df[0].loc[drop:]["week_end_date"])


dates_to_drop
```

```python
# Reset all indices and drop unwanted dates for relevant dataframes

# Empty array to hold cleaned dataframes is initialized
cleaned_df_array = []

# Find length of array holding dataframes
len_states = len(state_interest_df)

# Every dataframe in thearray of dataframes is cleaned identically
for i in range(len_states):
    
    # dataframe only containing wanted values is initialized, then the indices is reset with the result appended to the array
    #    of cleaned dataframes
    df = state_interest_df[i][state_interest_df[i]["week_end_date"].isin(dates_to_drop) == False]
    df.reset_index(inplace=True)
    df.drop("index", axis=1, inplace = True)
    cleaned_df_array.append(df)
    

cleaned_df_array[0]
```
